## Import libraries

```python
import requests
import os
```

## Read ACLED_key file

This file should contain the users key and email. For deatils consult the ACLED website and the [data export tool](https://acleddata.com/data-export-tool/) page.

```python
path = os.getcwd()
with open(path+'/ACLED_key.txt') as f:
    contents = f.readlines()
```

## URL used to request from ACLED API



```python
URL = 'https://api.acleddata.com/acled/read?key='+contents[1]+'&email='+contents[0]
```

## Filters

Search of events after a certain date and will more than 100 fatalities.

```python
EVENT_DATE = '&event_date=2022-04-01' #'&event_date={2022-04-01|2022-04-08}&event_date_where=BETWEEN'
FATALITIES = '&fatalities=100&fatalities_where=>'
```

## Make request with chosen filters

Warning: requests are limited.

```python
response = requests.get(URL+EVENT_DATE+FATALITIES)
```
